```python
import pandas as pd
from functools import reduce

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks  # dictionary {subject: marks}
    
    def total(self):
        return sum(self.marks.values())
    
    def average(self):
        return self.total() / len(self.marks)
    
    def grade(self):
        avg = self.average()
        if avg >= 90:
            return 'A'
        elif avg >= 75:
            return 'B'
        elif avg >= 60:
            return 'C'
        else:
            return 'D'

students = []

try:
    n = int(input("Enter number of students: "))
    for _ in range(n):
        name = input("Enter student name: ")
        subjects = ("Math", "Science", "English")
        marks = {sub: int(input(f"Enter marks for {sub}: ")) for sub in subjects}
        students.append(Student(name, marks))
except Exception as e:
    print("Error:", e)

# Using pandas for tabular output
data = {
    "Name": [s.name for s in students],
    "Total": [s.total() for s in students],
    "Average": [s.average() for s in students],
    "Grade": [s.grade() for s in students]
}

df = pd.DataFrame(data)
print("\nStudent Report Card:")
print(df)

# Find top student using reduce
top_student = reduce(lambda a, b: a if a.total() > b.total() else b, students)
print(f"\nTop Student: {top_student.name} with {top_student.total()} marks")
```
