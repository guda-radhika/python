```python
def divide(a, b):
    if b == 0:
        raise ZeroDivisionError("Cannot divide by zero!")
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        raise TypeError("Both inputs must be numbers")
    return a / b

try:
    result = divide(10, 0)
except ZeroDivisionError as zde:
    print(f"Caught ZeroDivisionError: {zde}")
except TypeError as te:
    print(f"Caught TypeError: {te}")
```

#### custom exception

```python
class NegativeNumberError(Exception):
    """Raised when the input number is negative"""
    pass

def check_number(num):
    if num < 0:
        raise NegativeNumberError("Input cannot be negative!")
    return num

try:
    value = int(input("Enter a number: "))
    check_number(value)
    
except NegativeNumberError as e:
    print(f"Caught custom exception: {e}")
```

```python
class ValidationError(Exception):
    def __init__(self, field, message):
        self.field = field
        self.message = message
        super().__init__(f"{field}: {message}")

def validate_age(age):
    if age < 0 or age > 120:
        raise ValidationError("age", "Age must be between 0 and 120")

try:
    validate_age(150)
except ValidationError as e:
    print(f"Caught custom exception -> {e}")
```

```python
class TooShortError(Exception):
    pass

class TooLongError(Exception):
    pass

def check_username(username):
    if len(username) < 3:
        raise TooShortError("Username too short!")
    if len(username) > 12:
        raise TooLongError("Username too long!")

try:
    check_username("pakfjkldjfakdjfakldfj")
except TooShortError as e:
    print(e)
except TooLongError as e:
    print(e)
```
